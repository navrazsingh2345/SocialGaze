```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
df = pd.read_csv("data.csv")
```

```python
# No. of users v/s regions 
region_counts = df['region'].value_counts()
plt.figure(figsize=(6, 4))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("User Distribution Across Regions", fontsize=14)
plt.show()
```

```python
# real and fake accounts count 
account_type_counts = df['account_type'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(account_type_counts, labels=account_type_counts.index, autopct='%1.1f%%', startangle=140, colors=['#66c2a5', '#fc8d62'])
plt.title('Proportion of Real vs. Fake Accounts')
plt.axis('equal')  
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.barplot(x='region', y='activity_pm', data=df, estimator=sum,palette='viridis')
plt.title('Total Activity per Month by Region')
plt.xlabel('Region')
plt.ylabel('Total Activity per Month')
plt.show()
```

```python
# Show how likes vary between different account types.
plt.figure(figsize=(10, 5))
sns.boxplot(x='account_type', y='total_likes', data=df, palette='Accent')
plt.title('Likes Distribution by Account Type')
plt.xlabel('Account Type')
plt.ylabel('Total Likes')
plt.show()
```

```python
# Visualize the average number of posts by account type.
plt.figure(figsize=(10, 5))
sns.barplot(x='account_type', y='total_posts', data=df, estimator='mean', palette='Set1')
plt.title('Average Total Posts by Account Type')
plt.xlabel('Account Type')
plt.ylabel('Average Total Posts')
plt.show()
```

```python
profile_pic_counts = df['profile_pic'].value_counts()
labels = ['With Profile Pic', 'Without Profile Pic']

fig, ax = plt.subplots(figsize=(6, 4))
ax.pie(profile_pic_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#b75c48', '#1ccadc'], wedgeprops={'width': 0.3},)

# Title and display
plt.title("Profile Picture Ownership", fontsize=14, fontweight='bold')
plt.show()
```

```python
# Account Type vs Activity Level
plt.figure(figsize=(8, 5))
sns.countplot(x='account_type', hue='activity_level', data=df)
plt.title('Account Type vs Activity Level')
plt.xlabel('Account Type')
plt.ylabel('Count')
plt.legend(title='Activity Level')
plt.show()
```

```python
df['total_reach'] = df['total_likes'] + df['total_comments']

sns.barplot(data=df, x='real_account_type', y='total_reach', hue='region', ci=None)
plt.title("Total Reach by Account Type")
plt.xlabel("Account Type")
plt.ylabel("Total Reach (Likes + Comments)")
plt.legend(title="Region")
plt.show()
```

```python
# Analyze the proportion of account types (real/fake) in different regions.
sns.countplot(data=df, x='region', hue='account_type',palette='winter_r')
plt.title("Region-wise Distribution of Account Types")
plt.xlabel("Region")
plt.ylabel("Count")
plt.legend(title="Account Type")
plt.show()
```

```python
profile_pic_data = df.groupby(['profile_pic', 'account_type']).size().unstack()

profile_pic_data.plot(kind='pie',subplots=True,autopct='%1.1f%%',figsize=(10, 5),startangle=90,labels=['Without Profile Pic', 'With Profile Pic'],
    legend=False,colors=['#66c2a5', '#fc8d62'])

plt.suptitle("Profile Picture Ownership vs Account Type", fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
# Profile Pic vs Account Type
plt.figure(figsize=(8, 5))
sns.countplot(x='profile_pic', hue='account_type', data=df, palette='Set1') 
plt.title('Profile Pic vs Account Type')
plt.xlabel('Profile Pic (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.legend(title='Account Type')
plt.show()
```

```python
# Activity Level vs Region

# Create a bar plot using Plotly
fig = px.bar(df, x="region", color="activity_level", title="Activity Level by Region", 
             color_discrete_map={"High": "green", "Medium": "orange", "Low": "red"})

# Show the interactive plot
fig.show()
```
